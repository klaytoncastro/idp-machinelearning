```python
import pandas as pd

url = 'https://raw.githubusercontent.com/klaytoncastro/idp-machinelearning/main/decision-tree/winequality-merged.csv'
arquivo = pd.read_csv(url)
arquivo.head()
```

```python
arquivo['color'] = arquivo['color'].replace('red', 0)
arquivo['color'] = arquivo['color'].replace('white', 1)
arquivo.head()
```

```python
import numpy as np
arquivo['worst'] = np.where(arquivo['quality'] < 7, 1, 0)
arquivo.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular a matriz de correlao para as colunas numricas do DataFrame
corr = arquivo.select_dtypes('number').corr()

# Personalizar a paleta de cores (opcional)
custom_palette = sns.color_palette("RdBu_r", n_colors=50)

# Plotar o mapa de calor usando o Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap=custom_palette, fmt=".2f")
plt.title('Mapa de Calor da Matriz de Correlao')
plt.show()
```

```python
# Calcular a matriz de correlao
correlation_matrix = arquivo.corr()

# Selecionar as correlaes com a varivel alvo ('quality')
correlation_with_target = correlation_matrix['worst']

# Excluir a correlao da varivel alvo consigo mesma
correlation_with_target = correlation_with_target.drop('worst')

# Exibir os valores de correlao e nomes das variveis preditoras
#print("Correlao com a varivel alvo (worst):\n")
#print(correlation_with_target)

# Ordenar os valores de correlao em ordem decrescente
correlation_with_target_sorted = correlation_with_target.sort_values(ascending=False)

# Exibir os valores de correlao e nomes das variveis preditoras ordenados
print("Correlao com a varivel alvo (worst) - Ordenado:\n")
print(correlation_with_target_sorted)
```

```python
#arquivo = arquivo.drop(['alcohol', free sulfur dioxide', total sulfur dioxide'], axis=1) # Teste removendo vrias colunas com baixa correlao

arquivo = arquivo.drop(['quality'], axis=1) #Removendo a coluna 'quality' para no interferir nos clculos do modelo para predio da varivel dependente 'worst'

arquivo.head()
```

```python
# Exibir o percentual de cada classe
print(arquivo['worst'].value_counts(normalize=True) * 100)
```

```python
# Definindo os atributos da funo de aprendizagem
y = arquivo['worst']
X = arquivo.drop('worst',axis = 1)
```

```python
from sklearn.model_selection import train_test_split

# Definindo os conjuntos de treino e teste, onde x  o conjunto de atributos (features que so nossas variveis preditoras) e y  a varivel alvo.
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=55)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

#Otimizao dos Hiperparmetros para o Modelo
modelo = ExtraTreesClassifier(n_estimators=40, max_depth=20, min_samples_leaf=1, min_samples_split=2, max_features="sqrt", class_weight="balanced", random_state=42)

#Treinamento e Predio
modelo.fit(x_train, y_train)
y_pred = modelo.predict(x_test)
```

```python
resultado = modelo.score(x_test, y_test)
print ("Acurcia:", resultado)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

"""
Calculando e exibindo a matriz de confuso. A orientao padro  a seguinte:
[0,0]: Verdadeiros Negativos (VN) - Previses corretamente identificadas como negativas.
[0,1]: Falsos Positivos (FP) - Previses incorretamente identificadas como positivas.
[1,0]: Falsos Negativos (FN) - Previses incorretamente identificadas como negativas.
[1,1]: Verdadeiros Positivos (VP) - Previses corretamente identificadas como positivas.
"""

conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confuso:")
print(conf_matrix)
```

```python
"""
Calculando e exibindo as mtricas de classificao.
Se algumas classes tm muito mais amostras do que outras, isso pode influenciar o desempenho e confiabilidade do modelo.

O "support" refere-se  quantidade de ocorrncias da classe especfica no conjunto de dados, sendo til para verificar desbalanceamentos.
A "macro avg" calcula a mdia aritmtica das mtricas (preciso, recall, F1-score) para cada classe, sem considerar o nmero de instncias em cada classe (support).
A "weighted avg" calcula a mdia ponderada das mtricas para cada classe, considerando o nmero de instncias em cada classe (support).
"""
print("Relatrio de Classsificao:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Prever as probabilidades para os dados de teste.
# Note que estamos interessados nas probabilidades da classe positiva (1), ento usamos [:, 1].
y_probs = modelo.predict_proba(x_test)[:, 1]

# Calcular FPR, TPR, e limiares
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Calcular a AUC
roc_auc = auc(fpr, tpr)

# Plotar a curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```
